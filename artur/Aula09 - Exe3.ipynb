### Exerc√≠cio 3

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

caminhoArquivo = './Arquivos/company_sales_data.csv'

df = pd.read_csv(caminhoArquivo)

df.head(1)

listaMes = df['month_number'].to_list()
faceCream = df['facecream'].to_list()
faceWash = df['facewash'].to_list()
toothPaste = df['toothpaste'].to_list()
bathingSoap = df['bathingsoap'].to_list()
shampooSales = df['shampoo'].to_list()
moisturizerSales = df['moisturizer'].to_list()

# Gerar um gr√°fico de linhas para plotar cada vari√°vel descrita acima
# Tipo  de marcador = o | linewidth = 3

plt.plot(listaMes, faceCream, label = 'Face Cream', marker = 'o', linewidth = 3)
plt.plot(listaMes, faceWash, label = 'Face Wash', marker = 'o', linewidth = 3)
plt.plot(listaMes, toothPaste, label = 'Tooth Paste', marker = 'o', linewidth = 3)
plt.plot(listaMes, bathingSoap, label = 'Bathing Soap', marker = 'o', linewidth = 3)
plt.plot(listaMes, shampooSales, label = 'Shampoo', marker = 'o', linewidth = 3)
plt.plot(listaMes, moisturizerSales, label = 'Moisturizer', marker = 'o', linewidth = 3)


# Adicionando legendas ao gr√°fico
plt.ylabel('Sales unit number')
plt.xlabel('Month Number')
plt.legend(loc = 'upper left')

#Ajustando os r√≥tulos do eixo
plt.xticks(listaMes)
plt.yticks([1000,2000,4000,6000,8000,10000,12000,14000,16000,18000])

# T√≠tulo do gr√°fico
plt.title('Dados de Venda')

# Exibir o gr√°fico
plt.show()
```

### Exerc√≠cio 04

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

caminhoArquivo = './Arquivos/company_sales_data.csv'

df = pd.read_csv(caminhoArquivo)

df.head(1)

# Gerar listas
listaMes = df['month_number'].to_list()
VendasPasta = df['toothpaste'].to_list()

# Criar o gr√°fico de dispers√£o com seaborn
sns.scatterplot(x = listaMes, y = VendasPasta, label = 'Vendas pasta de dente', c = 'red')

plt.title('Gr√°fico Dispers√£o')
plt.xlabel('M√™s')
plt.ylabel('Vendas')
plt.grid(True, linewidth = 0.5, linestyle = '--')
plt.xticks(listaMes)

```
