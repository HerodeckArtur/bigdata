### Trabalhando com Dataframe

```python
# Importar a bliblioteca
import pandas as pd

# Caminho do arquivo --> importante criar uma variável
caminho = "C:\\Users\\FIC\\Documents\\ArturHerodeck-python\\Arquivos\\bank-full.xlsx"

# Carregar o arquivo para um Dataframe
df = pd.read_excel(caminho, sheet_name='Sheet 1 - bank-full')

#Sheet 1 - bank-full --> tem que determinar a aba do excel
```

### Renomeando as colunas do Dataframe

```python
# Modificar os nomes das colunas
# As colunas devem estar na mesma ordem dos nomes
# A quantidade de nomes deve ser igual a quantidade de colunas

df.columns = [
    'Idade','Job','Estado civil','Educação','Default','Balance','Housing','Empréstimo','Contato','Dia','Mês','Duração','Campanha','Pdays','Previous','Poutcome'
]

df.head()
```

### Método df.rename()

```python
df = df.rename(columns={
    'Idade': 'IDADE',
    'Job':'Profissão'
})

df.head()
```
