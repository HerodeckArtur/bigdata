```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

caminhoArquivo = './Arquivos/company_sales_data.csv'

df = pd.read_csv(caminhoArquivo)

# Mostrar cabeçalho com uma linha
df.head(1)

# Gerar listas
cremeFacial = df['facecream'].to_list()
sabaoFacial = df['facewash'].to_list()
listaMes = df['month_number'].to_list()

# sns.barplot(x = listaMes, y = cremeFacial, label = 'Creme Facial') ---> Utilizando Seaborn 
# sns.barplot(x = listaMes, y = sabaoFacial, label = 'Sabão Facial') ---> Utilizando Seaborn

plt.bar([pos - 0.2 for pos in listaMes], cremeFacial, width = 0.4)
plt.bar([pos + 0.2 for pos in listaMes], sabaoFacial, width = 0.4)
plt.grid(True, linewidth = 0.5, linestyle = '--')
plt.xticks(listaMes)
```
