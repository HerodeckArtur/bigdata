```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression

# Importando o dataset
caminhoArquivo = ('./Arquivos/IceCreamData.csv')
df = pd.read_csv(caminhoArquivo)
```

```python
# Visualizar últimas linhas
df.tail(3)
```

```python
# Descrever os dados
df.describe()
```

```python
# Informações dos tipode de dados
df.info()
```

```python
# Gerar algumas visualizações gráficas
sns.jointplot(data= df, x= 'Temperature', y= 'Revenue', color= 'gray')
```

```python
sns.pairplot(df)
```

```python
sns.lmplot(x= 'Temperature', y= 'Revenue', data = df)
```

```python
# Criar e testar o modelo
# Criação da variável dependente (Y) --> resposta

y = df['Revenue']

# Criação da variável independente (X) --> entrada
X = df[['Temperature']]

# Separar os dados para teste e treino
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)

# Criação do modelo
modelo = LinearRegression()

# Treinamento do modelo
modelo.fit(X_train,y_train)

# Exibição dos coeficientes

print('Coeficiente linear (b): ', modelo.intercept_)
print('Coeficiente angular (a): ', modelo.coef_)
```

```python
# Testando o modelo
y_predict = modelo.predict(X_test)
y_predict
```
