### Primeiros passos com Matplotlib

```python
# Importando as bibliotecas

import matplotlib.pyplot as plt
import numpy as np

# Criando os arrays (matrizes) para utilizarnnos gráficos

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Criando um gráfico com os valores anteriores
plt.plot(x,y)
```

### Modificando o tipo de marcador

```python
plt.plot(x, y, marker = 'X', c = 'red')
```

### Modificando a forma como os pontos são ligados e o tipo de linha

```python
plt.plot(
    x,
    y,
    marker = '*',
    c = 'blue',
    ds = 'steps', # Forma de ligação
    ls = '--' # Estilo da linha
)
```

### Alterando as cores dos marcadores

```python
plt.plot(
    x, # Vetor responsável pelos valores do eixo X
    y, # Veotr responsável pelos valores do eixo Y
    marker = '*', # Marcador do gráfico
    c = 'red', # Cor da linha
    ds = 'steps', # Forma de ligação
    ls = '--', # Estilo da linha
    mec = 'blue' # Cor dos marcadores

)
```

### Adicionando outros elementos ao gráfico

```python
# Criando o gráfico
plt.plot(x, y)

# Adicionando o título do eixo X
plt.xlabel('Valores de X', fontsize = 12)

# Adicionando o título do eixo Y
plt.ylabel('Valores de Y', fontsize = 12)

# Adicionando o título do gráfico
plt.title('Título do gráfico', fontsize = 20, c = 'red')

# Alterando as cores dos rótulos de x e y
plt.xticks(fontsize = 8, c = 'blue')
plt.yticks(fontsize = 8, c = 'purple')
```

```python
# Importar as blbliotecas

import pandas as pd
import matplotlib.pyplot as plt
import os

# Mapear o caminho do arquivo

caminhoArquivo = './company_sales_data.csv'

# Alocar os dados em um dataframe
df = pd.read_csv(caminhoArquivo)

# Visualizar o cabeçalho do Dataframe
df.head(2)
```

```python
# Selecionar os dados que vão ser apresentados no gráfico
listaLucro = df['total_profit'].to_list()
listaMeses = df['month_number'].to_list()

# Criar o gráfico utilizando os dados listados aanteriormente
plt.plot(listaMeses, listaLucro, label = 'Lucro mensal do ano anterior')

plt.xlabel('Número do mês') # Título do eixo X
plt.ylabel('Lucro') # Título do eixo Y
plt.title('Lucro mensal da empresa') # Título do gráfico
plt.legend() # Legenda do gráfico

```
